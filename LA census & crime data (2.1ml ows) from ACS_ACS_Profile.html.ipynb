```python
## scraping LA city crime data
## Data Title "Crime Data from 2010 to 2019" for LA
## website: https://data.lacity.org/Public-Safety/Crime-Data-from-2010-to-2019/63jg-8b9z  2.21 ml rows

## Example from Baltimore: https://data.baltimorecity.gov/resource/wsfq-mvij.json  
import requests
import pandas as pd

#2.1 ml retrieved
data = requests.get('https://data.lacity.org/resource/63jg-8b9z.json?$limit=2000000&$select=LOCATION,LAT,LON')
##  CRS, WGS 1984, decimal degrees (EPSG 4326)

LA_Crime_df = pd.DataFrame(data=data.json())
```

```python
## save LA city tract file to local drive - NOT ENOUGH MEMORY FOR 2.2ml rows in Excel
##dc_NYCAll_withCrimeNoNAs.to_csv("dc_NYCAll_withCrimeNov30th.csv")
LA_Crime_df.to_csv("LA_crime_as_downloaded_Feb9th2021.csv") ## 122MB
```

```python
LA_Crime_df['LON'].isnull().sum() ## there are None

##df = df[df['EPS'].notna()] to keep rows that have not got NAs

LA_Crime_df = LA_Crime_df[LA_Crime_df['LON'].notna()]
print("No of NAs in longitude col", LA_Crime_df['LON'].isnull().sum())
print("shape of LA_Crime_df", LA_Crime_df.shape)  
LA_Crime_df
```

```python
import geopandas as gpd

## Creating a GeoDataframe file from LA_crime_df and then using it  
## to get tract and block ID points from the California (06) census shape file 

LA_data_points = gpd.GeoDataFrame(LA_Crime_df, geometry=gpd.points_from_xy(LA_Crime_df.LON, LA_Crime_df.LAT), crs='EPSG:4269')
LA_data_points

LA_shape = "cb_2018_06_tract_500k.shp"  # load California state code "06" for LA shape file 

LA_map = gpd.read_file(LA_shape)

LA_points_in_map = gpd.sjoin(LA_data_points.to_crs('EPSG:4269'), 
                         LA_map, 
                         how='inner', 
                         op='within')  ## doing inner because the census data is for whole of California and only LA crime data census tracts
print("I am done")
print("Today's date is:", date.today())
```
