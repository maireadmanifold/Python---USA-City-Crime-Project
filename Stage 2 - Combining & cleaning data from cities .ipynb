```python
## Step 2 joins data from various city and then cleans up the data, turning all data to numeric

import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 100)
url_NYC = 'https://raw.githubusercontent.com/maireadmanifold/files/main/NYC_CSV__Feb21th_before.csv'
df_ML_NYC = pd.read_csv(url_NYC, error_bad_lines=False, index_col=0)
url_Baltimore = 'https://raw.githubusercontent.com/maireadmanifold/files/main/Baltimore_CSV__Feb8th_before.csv'
df_ML_Baltimore = pd.read_csv(url_Baltimore, error_bad_lines=False, index_col=0)
url_Chicago = 'https://raw.githubusercontent.com/maireadmanifold/files/main/Chicago_CSV__Feb12th_before.csv'
df_ML_Chicago = pd.read_csv(url_Chicago, error_bad_lines=False, index_col=0)
url_LA = 'https://raw.githubusercontent.com/maireadmanifold/files/main/LA_CSV__Feb9th_before.csv'
df_ML_LA = pd.read_csv(url_LA, error_bad_lines=False, index_col=0)
url_SF = 'https://raw.githubusercontent.com/maireadmanifold/files/main/SF_CSV__Feb8th_before.csv'
df_ML_SF = pd.read_csv(url_SF, error_bad_lines=False, index_col=0)
url_Dallas = 'https://raw.githubusercontent.com/maireadmanifold/files/main/Dallas_CSV_Feb8th_before.csv'
df_ML_Dallas = pd.read_csv(url_Dallas, error_bad_lines=False, index_col=0)

```

```python
data_frames = [df_ML_NYC, df_ML_Baltimore, df_ML_Chicago, df_ML_LA, df_ML_SF, df_ML_Dallas]
All_CSVs = pd.concat(data_frames)
```

```python
## REMOVE when done with individual data files and adding in index from shape file
All_CSVs.to_csv("All_Step2_CSVsMar6th.csv")
```

```python
## changing everything to numeric 
import numpy as np
All_CSVs_0 = All_CSVs.copy()
All_CSVs_0[
    All_CSVs_0.apply(lambda col: pd.to_numeric(col, errors='coerce')).apply(lambda x: x < 0)
] = np.NaN
print("No of rows", All_CSVs_0.shape)
All_CSVs_0.head(4)
```

```python
#DP03_0065E-HH-MeanEarnings
#groupedTract_count = Dallas_points_in_map.groupby(['County','Tract','AreaLand']).size().reset_index(name='crimeStats')
#CountyTract_MeanIncome = All_CSVs.groupby(['County', 'Tract']).mean().reset_index(name='MeanEarnings')
#All_CSVs['%ofTractCountyMeanEarning']
groupedStateCountyMeanEarning = All_CSVs_0.groupby(['State','County'])['DP03_0065E-HH-MeanEarnings'].mean().sort_values(ascending=False).reset_index(name="meanCountyStateEarning")
groupedStateCountyWorkersMeanEarnings = All_CSVs_0.groupby(['State','County'])['DP03_0092E-Median Earning Workers'].mean().sort_values(ascending=False).reset_index(name="meanCountyStateWorkerEarnings")
groupedStateCountyMaleMeanEarnings = All_CSVs_0.groupby(['State','County'])['DP03_0093E-Median Earnings Male FT Workers'].mean().sort_values(ascending=False).reset_index(name="meanCountyStateMaleEarnings")
groupedStateCountyFemaleMeanEarnings = All_CSVs_0.groupby(['State','County'])['DP03_0094E-Median Earnings Female FT Workers'].mean().sort_values(ascending=False).reset_index(name="meanCountyStateFemaleEarnings")
groupedStateCountyDP040101EMedMortgage = All_CSVs_0.groupby(['State','County'])['DP04_0101E-Med Onwer Monthly Mortgage'].mean().sort_values(ascending=False).reset_index(name="meanCountyStateOwnerMortgage")
groupedStateCountyDP04_0134EMedGrossRent = All_CSVs_0.groupby(['State','County'])['DP04_0134E-Median Gross Rent'].mean().sort_values(ascending=False).reset_index(name="meanCountyStateRent")
groupedStateCountyDP05_0070PEHandL_Race = All_CSVs_0.groupby(['State','County'])['DP05_0070PE-%H&L race'].mean().sort_values(ascending=False).reset_index(name="meanCountyStateHandL_Race") # Hispanic and Latino
## added in later on
groupedStateCountyDP04_0045PEAvgHousingUnits = All_CSVs_0.groupby(['State','County'])['DP04_0045PE-%Occup HU'].mean().sort_values(ascending=False).reset_index(name="meanCountyStateAvgNoHU")
groupedStateCountyDP02_0015EAvgHHSize = All_CSVs_0.groupby(['State','County'])['DP02_0015E-Avg HH Size'].mean().sort_values(ascending=False).reset_index(name="meanCountyStateAvgHHSize")
groupedStateCountyDP04_0016PEAvgFamSize = All_CSVs_0.groupby(['State','County'])['DP02_0016E-Avg Fam Size'].mean().sort_values(ascending=False).reset_index(name="meanCountyState_AvgFamSize")
```

```python
All_CSVs_1 = pd.merge(left = All_CSVs_0, right = groupedStateCountyMeanEarning, how = 'inner', on = ['State', 'County'])

#All_CSVs_1[All_CSVs_1['Tract']==19700]
```

```python
All_CSVs_1['%DP03_0065E-HH-MeanEarnings'] = (All_CSVs_1['DP03_0065E-HH-MeanEarnings']/All_CSVs_1['meanCountyStateEarning']*100) #).apply(lambda x: '{pct_mean_earnings:.1%}'.format(pct_mean_earnings = x))
```

```python
All_CSVs_2 = pd.merge(left = All_CSVs_1, right = groupedStateCountyWorkersMeanEarnings, how = 'inner', on = ['State', 'County'])
```

```python
All_CSVs_2['%DP03_0092E-Median Earning Workers'] = (All_CSVs_2['DP03_0092E-Median Earning Workers']/All_CSVs_2['meanCountyStateWorkerEarnings']*100) #).apply(lambda x: '{pct_mean_earnings:.1%}'.format(pct_mean_earnings = x))
```

```python
All_CSVs_3 = pd.merge(left = All_CSVs_2, right = groupedStateCountyMaleMeanEarnings, how = 'inner', on = ['State', 'County'])
```

```python
All_CSVs_3['%DP03_0093E-Median Earnings Male FT Workers'] = (All_CSVs_3['DP03_0093E-Median Earnings Male FT Workers']/All_CSVs_3['meanCountyStateMaleEarnings']*100) #).apply(lambda x: '{pct_mean_earnings:.1%}'.format(pct_mean_earnings = x))
```

```python
All_CSVs_4 = pd.merge(left = All_CSVs_3, right = groupedStateCountyFemaleMeanEarnings, how = 'inner', on = ['State', 'County'])
```

```python
All_CSVs_4['%DP03_0094E-Median Earnings Female FT Workers'] = (All_CSVs_4['DP03_0094E-Median Earnings Female FT Workers']/All_CSVs_4['meanCountyStateFemaleEarnings']*100) #).apply(lambda x: '{pct_mean_earnings:.1%}'.format(pct_mean_earnings = x))
```

```python
All_CSVs_5 = pd.merge(left = All_CSVs_4, right = groupedStateCountyDP040101EMedMortgage, how = 'inner', on = ['State', 'County'])
```

```python
All_CSVs_5['%DP04_0101E-Med Onwer Monthly Mortgage'] = (All_CSVs_5['DP04_0101E-Med Onwer Monthly Mortgage']/All_CSVs_5['meanCountyStateOwnerMortgage']*100) #.apply(lambda x: '{pct_mean_earnings:.1%}'.format(pct_mean_earnings = x))
```

```python
All_CSVs_6 = pd.merge(left = All_CSVs_5, right = groupedStateCountyDP04_0134EMedGrossRent, how = 'inner', on = ['State', 'County'])
```

```python
All_CSVs_6['%DP04_0134E-Median Gross Rent'] = (All_CSVs_6['DP04_0134E-Median Gross Rent']/All_CSVs_6['meanCountyStateRent']*100) #).apply(lambda x: '{pct_mean_earnings:.1%}'.format(pct_mean_earnings = x))
```

```python
All_CSVs_7 = pd.merge(left = All_CSVs_6, right = groupedStateCountyDP05_0070PEHandL_Race, how = 'inner', on = ['State', 'County'])
```

```python
All_CSVs_7['%DP05_0070PE-%H&L race'] = (All_CSVs_7['DP05_0070PE-%H&L race']/All_CSVs_7['meanCountyStateHandL_Race']*100) #.apply(lambda x: '{pct_mean_earnings:.1%}'.format(pct_mean_earnings = x))
```

```python
## Next is No Housing units in tract, Avg Housing Size and Avg Family size where we want % of tract value compared to avg Housing and Family size.  
```

```python
All_CSVs_8 = pd.merge(left = All_CSVs_7, right = groupedStateCountyDP04_0045PEAvgHousingUnits, how = 'inner', on = ['State', 'County'])
```

```python
All_CSVs_8['%DP04_0045PE-%Occup HU'] =  (All_CSVs_8['DP04_0045PE-%Occup HU']/All_CSVs_8['meanCountyStateAvgNoHU']*100) #.apply(lambda x: '{pct_mean_earnings:.1%}'.format(pct_mean_earnings = x))
```

```python
## Onto merge 9 here, so SKIPPED 8  )
All_CSVs_9 = pd.merge(left = All_CSVs_8, right = groupedStateCountyDP02_0015EAvgHHSize, how = 'inner', on = ['State', 'County'])
```

```python
All_CSVs_9['%DP02_0015E-Avg HH Size'] = (All_CSVs_9['DP02_0015E-Avg HH Size'] /All_CSVs_9['meanCountyStateAvgHHSize']*100)
```

```python
## need merge here
All_CSVs_10 = pd.merge(left = All_CSVs_9, right = groupedStateCountyDP04_0016PEAvgFamSize, how = 'inner', on = ['State', 'County'])
```

```python
All_CSVs_10['%DP02_0016E-Avg Fam Size'] =  (All_CSVs_10['DP02_0016E-Avg Fam Size']/All_CSVs_10['meanCountyState_AvgFamSize']*100) 
```

```python
#######REMOVING ROWS  

## Removing population density columns after creating subset dataframe "Popo_Density_Per_Hector"
## Removing MeanEarning columns after creating for DP03_0065E % calcualtion
All_CSVs_11 = All_CSVs_10
All_CSVs_11.drop('Popo_Density_Per_Hector', inplace = True, axis=1)
All_CSVs_11.drop('AreaLand',inplace = True, axis=1)
All_CSVs_11.drop('ALAND', inplace = True, axis=1)

## Removing total population and total family households after creating a subset dataframe "TotalPopoandTotalFamHHs"
#'DP05_0001E-T Popo', 'DP02_0002E-T Fam HHs'
All_CSVs_11.drop('DP05_0001E-T Popo', inplace = True,  axis=1)
All_CSVs_11.drop('DP02_0002E-T Fam HHs', inplace = True,  axis=1)



#removing DP columns which were estimates and where mean was used to calculate % mean for state county combination 
All_CSVs_11.drop('DP03_0065E-HH-MeanEarnings', inplace = True,  axis=1)
All_CSVs_11.drop('meanCountyStateEarning', inplace = True,  axis=1)
All_CSVs_11.drop('DP03_0092E-Median Earning Workers', inplace = True,  axis=1)
All_CSVs_11.drop('meanCountyStateWorkerEarnings', inplace = True,  axis=1)
All_CSVs_11.drop('DP03_0093E-Median Earnings Male FT Workers', inplace = True,  axis=1)
All_CSVs_11.drop('meanCountyStateMaleEarnings', inplace = True,  axis=1)
All_CSVs_11.drop('DP03_0094E-Median Earnings Female FT Workers', inplace = True,  axis=1)
All_CSVs_11.drop('meanCountyStateFemaleEarnings', inplace = True,  axis=1)
All_CSVs_11.drop('DP04_0101E-Med Onwer Monthly Mortgage', inplace = True,  axis=1)
All_CSVs_11.drop('meanCountyStateOwnerMortgage', inplace = True,  axis=1)
All_CSVs_11.drop('DP04_0134E-Median Gross Rent', inplace = True,  axis=1)
All_CSVs_11.drop('meanCountyStateRent', inplace = True,  axis=1)
All_CSVs_11.drop('DP05_0070PE-%H&L race', inplace = True,  axis=1)
All_CSVs_11.drop('meanCountyStateHandL_Race', inplace = True,  axis=1)

All_CSVs_11.drop('DP04_0006E-T HU', inplace = True,  axis=1)## remove DP04_0006E-T HU - Another Total HU - don't need
All_CSVs_11.drop('DP04_0001E-T H U', inplace = True,  axis=1)
#Remove DP04_0045PE-%Occup HU and DP02_0001E-T HHs now that %DP04_0045PE-%Occup HU exists 
All_CSVs_11.drop('DP04_0045PE-%Occup HU', inplace = True,  axis=1)
All_CSVs_11.drop('DP02_0001E-T HHs', inplace = True,  axis=1)
#DP04_0048E-Owner Ave Household size - don't understand this output
All_CSVs_11.drop('DP04_0048E-Owner Ave Household size', inplace = True,  axis=1)

## DP02_0015E and DP02_0016E Avg HH Size and Avg Fam Size removing because now have % of HH and Family as per state and county
All_CSVs_11.drop('DP02_0015E-Avg HH Size', inplace = True,  axis=1)
All_CSVs_11.drop('DP02_0016E-Avg Fam Size', inplace = True,  axis=1)
```

```python
# drop meanCountyStateAvgHHSize and meanCountyState_AvgFamSize
All_CSVs_11.drop('meanCountyStateAvgHHSize', inplace = True,  axis=1)
All_CSVs_11.drop('meanCountyState_AvgFamSize', inplace = True,  axis=1)
```
