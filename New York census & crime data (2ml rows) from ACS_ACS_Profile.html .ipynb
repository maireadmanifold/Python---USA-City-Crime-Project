```python
## Scraping New York crime data
## json api https://data.cityofnewyork.us/resource/sv2w-rv3k.json

## website: https://data.cityofnewyork.us/Public-Safety/NYPD-Criminal-Court-Summons-Historic-/sv2w-rv3k  5.28m rows 21/2/2021 

import requests
import pandas as pd
#1ml retrieved
#data = requests.get('https://data.cityofnewyork.us/resource/sv2w-rv3k.json?$limit=1000000&$select=BORO,latitude,longitude').json()
#10k retrieved
data = requests.get('https://data.cityofnewyork.us/resource/sv2w-rv3k.json?$limit=5300000&$select=BORO,latitude,longitude').json()

##  CRS, WGS 1984, decimal degrees (EPSG 4326)

NYC_Crime_df = pd.DataFrame(data=data)
NYC_Crime_df.to_csv("NewYork_crime_as_downloaded_Feb21th2021.csv") ## 5.2m rows 21/2/2021
NYC_Crime_df.head() 
```

```python
NYC_Crime_df['longitude'].isnull().sum() ## there are None

##df = df[df['EPS'].notna()] to keep rows that have not got NAs

NYC_Crime_df = NYC_Crime_df[NYC_Crime_df['longitude'].notna()]
print("No of NYC NAs in longitude col", NYC_Crime_df['longitude'].isnull().sum())
print("shape of NYC_Crime_df", NYC_Crime_df.shape)  
NYC_Crime_df
```

```python
import geopandas as gpd ## completing for tracts


nyc_data_points = gpd.GeoDataFrame(NYC_Crime_df, geometry=gpd.points_from_xy(NYC_Crime_df.longitude, NYC_Crime_df.latitude), crs='EPSG:4326')

nyc_shape = "cb_2018_36_tract_500k.shp"  # load shape file 

nyc_map = gpd.read_file(nyc_shape)

NYC_points_in_map = gpd.sjoin(nyc_data_points.to_crs('EPSG:4269'), 
                         nyc_map, 
                         how='inner', 
                         op='within')
print("I am done")
print("Today's date is:", date.today())
```
