```python
#NYC data 
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
url_NYC = 'https://raw.githubusercontent.com/maireadmanifold/files/main/NYC_CSV_Cleaned_and_no_shape_Mar8th.csv'
df_ML_NYC = pd.read_csv(url_NYC, error_bad_lines=False, index_col=0)
```

```python
df_ML_NYC[1:2]
```

```python
print("shape NYC: ", df_ML_NYC.shape)
```

```python
df_ML_NYC['COUNTYFP'] = df_ML_NYC['Countyfp'].apply(lambda x: '{0:0>3}'.format(x)).astype(str)
df_ML_NYC['TRACTCE'] = df_ML_NYC['Tractce'].apply(lambda x: '{0:0>6}'.format(x)).astype(str)
df_ML_NYC['STATEFP'] = df_ML_NYC['Statefp'].apply(lambda x: '{0:0>2}'.format(x)).astype(str)
df_ML_NYC.drop('Statefp', inplace = True,  axis=1)
df_ML_NYC.drop('Countyfp', inplace = True,  axis=1)
df_ML_NYC.drop('Tractce', inplace = True,  axis=1)
```

```python
df_ML_NYC[1:2]
```

```python
## opened a copy of cb_2018_36_tract_500k.dbf in Excel and save as a csv file. (36 is code for New York)
## the index_right gives aligns with index in csv file. 
import pandas as pd
census_tract_NYC = pd.read_csv("cb_2018_36_tract_500k.csv", header=0)
print("census tract info from New York dbf file saved as a csv file")
census_tract_NYC[1:2]
```

```python
census_tract_NYC['COUNTYFP'] = census_tract_NYC['COUNTYFP'].apply(lambda x: '{0:0>3}'.format(x)).astype(str)
census_tract_NYC['TRACTCE'] = census_tract_NYC['TRACTCE'].apply(lambda x: '{0:0>6}'.format(x)).astype(str)
census_tract_NYC['STATEFP'] = census_tract_NYC['STATEFP'].apply(lambda x: '{0:0>2}'.format(x)).astype(str)
```

```python
#myDataFrame.set_index('column_name')
census_tract_NYC.reset_index(inplace=True)
census_tract_NYC[1:2]
```

```python
##points_in_map_withTrees = pd.merge(left = points_in_map, right = groupedTract_count, how = 'inner', on = ['COUNTYFP', 'TRACTCE', 'ALAND'] )
df_ML_NYC_with_index_right = pd.merge(left = df_ML_NYC, right = census_tract_NYC, on=['STATEFP','COUNTYFP','TRACTCE'])
```

```python
df_ML_NYC_with_index_right[1:3]  ## with index from shape file included
```

```python
df_ML_NYC_with_index_right.drop('NAME_y', inplace = True,  axis=1)
df_ML_NYC_with_index_right.drop('AWATER', inplace = True,  axis=1)
df_ML_NYC_with_index_right.drop('LSAD', inplace = True,  axis=1)
df_ML_NYC_with_index_right.drop('ALAND', inplace = True,  axis=1)
#df2 = df.rename({'a': 'X', 'b': 'Y'}, axis=1) 
```

```python
df_ML_NYC_with_index_right.rename({'NAME_x': 'NAME'}, axis = 1, inplace = True) 
df_ML_NYC_with_index_right[1:2]
```

```python
df_ML_NYC_with_index_right.to_csv("NYC_CSV_Mar8th_after_shape.csv") ## for use in Stage 5 for use Tableau
df_ML_NYC_with_index_right.to_excel("NYC_CSV_Mar8th_after_shape.xlsx") ## for use in Stage 5 for use in Tableau
```
