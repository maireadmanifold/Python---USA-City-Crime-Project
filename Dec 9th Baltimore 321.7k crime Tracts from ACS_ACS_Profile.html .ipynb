```python
## scraping Baltimore city crime data
## Crimes data title: "Part 1 Crime data"
## Website: https://data.baltimorecity.gov/datasets/part-1-crime-data-3/data

## Baltimore API changed from .json in early 2020 to .geojson in Dec 2020
## original .json api was: https://data.baltimorecity.gov/resource/wsfq-mvij.json
import geopandas as gpd
import requests
import pandas as pd

## see link about reading in: https://stackoverflow.com/questions/42753745/how-can-i-parse-geojson-with-python
data2 = gpd.read_file('https://opendata.arcgis.com/datasets/acad702852104368a2b5fa0300e6661a_0.geojson?&$select=Location,Neighborhood,Latitude,Longitude')

##  CRS, WGS 1984, decimal degrees (EPSG 4326)

#BA_Crime_df = pd.DataFrame(data=data2.json())
#BA_Crime_df.head() ## 321k crime stats downloaded for BA  
data2.to_csv("Baltimore_crime_as_downloaded_Feb8th2021.csv")
data2.head()
```

```python
names = ['Location', 'Latitude','Longitude', 'GeoLocation']
BA_Crime_df = data2[names]
BA_Crime_df.head()
```

```python
## Noticed there were some NAs int he Longitude column - found 313 NA rows. 
BA_Crime_df['Longitude'].isnull().sum() ## there are None

##df = df[df['EPS'].notna()] to keep rows that have not got NAs

BA_Crime_df = BA_Crime_df[BA_Crime_df['Longitude'].notna()]
print("No of NAs in longitude col", BA_Crime_df['Longitude'].isnull().sum())
print("shape of BA_Crime_df", BA_Crime_df.shape)  
BA_Crime_df
```

```python
import geopandas as gpd

## Creating a GeoDataframe file from BA_crime_df and then using it  
## to get tract points from the Maryland census shape file 

BA_data_points = gpd.GeoDataFrame(BA_Crime_df, geometry=gpd.points_from_xy(BA_Crime_df.Longitude, BA_Crime_df.Latitude), crs='EPSG:4269')
BA_data_points

BA_shape = "cb_2018_24_tract_500k.shp"  # load Maryland state code "24" for BA (Baltimore) shape file 

BA_map = gpd.read_file(BA_shape)

BA_points_in_map = gpd.sjoin(BA_data_points.to_crs('EPSG:4269'), 
                         BA_map, 
                         how='inner', 
                         op='within')  ## doing inner because the census data is for whole of California and only BA crime data census tracts
print("I am done")
```

```python
import pandas as pd
## saved the dcb_2018_tract_500k.dbf file as a csv file to get access to the tract ID
## the index_right gives aligns with the tract ID in this file. 
## Remember that Maryland state ID = "24"
census_tract_df_Maryland = pd.read_csv("cb_2018_24_tract_500k.csv", header=0)
print("census tract info from Maryland (Baltimore) dbf file saved as a csv file")
census_tract_df_Maryland 
```

```python
census_tract_df_Maryland['TRACTCE'] = census_tract_df_Maryland['TRACTCE'].apply(lambda x: '{0:0>6}'.format(x)).astype(str)
census_tract_df_Maryland['TRACTCE']
```

```python
census_tract_df_Maryland['COUNTYFP'] = census_tract_df_Maryland['COUNTYFP'].apply(lambda x: '{0:0>3}'.format(x)).astype(str)
census_tract_df_Maryland['COUNTYFP']

```

```python
##The points_in_map of crime data does not include the county or tract or land area, so using join to add these 
## fields from Shape dbf file to the crime data. 
#points_in_map = points_in_map.join(census_tract_df[['COUNTYFP']], on='index_right')
#points_in_map = points_in_map.join(census_tract_df[['TRACTCE']], on='index_right')
#points_in_map = points_in_map.join(census_tract_df[['ALAND']], on='index_right')
BA_points_in_map2 = BA_points_in_map.join(census_tract_df_Maryland[['STATEFP','COUNTYFP','TRACTCE','ALAND']],on='index_right', lsuffix='_left', rsuffix='_right')
```

```python
import pandas as pd
#'COUNTYFP_right', 'TRACTCE_right',, 'ALAND_right'
BA_points_in_map2.drop('STATEFP_right', inplace=True, axis=1)
BA_points_in_map2.drop('COUNTYFP_right', inplace=True, axis=1)
BA_points_in_map2.drop('TRACTCE_right', inplace=True, axis=1)
BA_points_in_map2.drop('ALAND_right', inplace=True, axis=1)
#BA_points_in_map.drop('STATEFP_left', inplace=True, axis=1)
#BA_points_in_map.drop('COUNTYFP_left', inplace=True, axis=1)
#BA_points_in_map.drop('TRACTCE_left', inplace=True, axis=1)
#BA_points_in_map.drop('ALAND_left', inplace=True, axis=1)

```

```python
groupedTract_count = BA_points_in_map2.groupby(['County','Tract','AreaLand']).size().reset_index(name='crimeStats')
#df2 = df.rename({'a': 'X', 'b': 'Y'}, axis=1) 
groupedTract_count
```

```python
groupedTract_mean = groupedTract_count["crimeStats"].mean()
print("mean",groupedTract_mean)
groupedTract_min = groupedTract_count["crimeStats"].min()
print("min", groupedTract_min)
groupedTract_max = groupedTract_count["crimeStats"].max()
print("max", groupedTract_max)
```

```python
import numpy as np
groupedTract_count["crime"] = np.where(groupedTract_count.crimeStats < groupedTract_mean, "low", "high")
groupedTract_count
```
