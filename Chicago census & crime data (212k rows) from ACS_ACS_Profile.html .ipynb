```python
## scraping Chicago city crime data
## Data title "Crimes: reported incidents"
## Website: https ://www.chicago.gov/city/en/dataset/crime.html
## Example this is LA's:  https://data.lacity.org/Public-Safety/Crime-Data-from-2010-to-2019/63jg-8b9z
## Example this is Baltimore's https://data.baltimorecity.gov/resource/wsfq-mvij.json 
## Example this is New York's https://data.cityofnewyork.us/resource/sv2w-rv3k.json
import requests
import pandas as pd



#10k retrieved
data = requests.get('https://data.cityofchicago.org/resource/x2n5-8w5q.json?$limit=250000&$select=block,longitude, latitude')
##  CRS, WGS 1984, decimal degrees (EPSG 4326)

Chicago_Crime_df = pd.DataFrame(data=data.json())
Chicago_Crime_df.head()
```

```python
## Noticed there were no NAs in the Longitude column -
Chicago_Crime_df['longitude'].isnull().sum() ## there are 1370 on Feb 8th 2021 (204,222 - 202,852)
print("No of NAs in longitude col", Chicago_Crime_df['longitude'].isnull().sum())
##df = df[df['EPS'].notna()] to keep rows that have not got NAs

Chicago_Crime_df = Chicago_Crime_df[Chicago_Crime_df['longitude'].notna()]
print("No of NAs in longitude col", Chicago_Crime_df['longitude'].isnull().sum())
print("shape of Chicago_Crime_df", Chicago_Crime_df.shape)  
Chicago_Crime_df.head()
print("No of rows: ", Chicago_Crime_df.shape) ## 202,581 on 9/2/2021, 1,422 were null longitude rows
```

```python
import geopandas as gpd

## Creating a GeoDataframe file from Chicago_crime_df and then using it  
## to get tract points from the Illinois (State Id: 17) census shape file 

Chicago_data_points = gpd.GeoDataFrame(Chicago_Crime_df, geometry=gpd.points_from_xy(Chicago_Crime_df.longitude, Chicago_Crime_df.latitude), crs='EPSG:4269')
Chicago_data_points

Chicago_shape = "cb_2018_17_tract_500k.shp"  # load Maryland state code "24" for Chicago (Chicagoltimore) shape file 

Chicago_map = gpd.read_file(Chicago_shape)

Chicago_points_in_map = gpd.sjoin(Chicago_data_points.to_crs('EPSG:4269'), 
                         Chicago_map, 
                         how='inner', 
                         op='within')  ## doing inner because the census data is for whole of California and only Chicago crime data census tracts
print("I am done")
print("Today's date is:", date.today())
```

```python
groupedTract_count = Chicago_points_in_map2.groupby(['County','Tract','AreaLand']).size().reset_index(name='crimeStats')
#df2 = df.rename({'a': 'X', 'b': 'Y'}, axis=1) 
groupedTract_count = groupedTract_count.rename({'longitude': 'LON', 'latitude':'LAT'}, axis = 1)
groupedTract_count
```

```python
groupedTract_mean = groupedTract_count["crimeStats"].mean()
print("mean",groupedTract_mean)
groupedTract_min = groupedTract_count["crimeStats"].min()
print("min", groupedTract_min)
groupedTract_max = groupedTract_count["crimeStats"].max()
print("max", groupedTract_max)
```

```python
import numpy as np
groupedTract_count["crime"] = np.where(groupedTract_count.crimeStats < groupedTract_mean, "low", "high")
groupedTract_count
```
