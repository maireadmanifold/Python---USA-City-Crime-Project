```python
## Scraping SF crime data
## Title of data: 'Police Department Incident Reports: Historical 2003 to May 2018'
## Website: https://data.sfgov.org/Public-Safety/Police-Department-Incident-Reports-Historical-2003/tmnf-yvry
## API: https://data.sfgov.org/resource/tmnf-yvry.json

## Example Baltimore: https://data.baltimorecity.gov/resource/wsfq-mvij.json - Baltimore 
import requests
import pandas as pd
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
#2.16 ml retrieved
data = requests.get('https://data.sfgov.org/resource/tmnf-yvry.json?$limit=2500000&$select=address, x,y')
##  CRS, WGS 1984, decimal degrees (EPSG 4326)

SF_Crime_df = pd.DataFrame(data=data.json())
SF_Crime_df.to_csv("SF_crime_as_downloaded_Feb8th2021.csv")
SF_Crime_df ## 321k crime stats downloaded for BA  
```

```python
#SF_Crime_df['SF_Crime_df'].isnull().sum() ## there are None

##df = df[df['EPS'].notna()] to keep rows that have not got NAs

SF_Crime_df = SF_Crime_df[SF_Crime_df['x'].notna()]
print("No of NAs in SF longitude col", SF_Crime_df['x'].isnull().sum())
print("shape of SF_Crime_df", SF_Crime_df.shape)  
SF_Crime_df
```

```python
import geopandas as gpd

## Creating a GeoDataframe file from LA_crime_df and then using it  
## to get tract and block ID points from the Illinois shape file from census 

SF_data_points = gpd.GeoDataFrame(SF_Crime_df, geometry=gpd.points_from_xy(SF_Crime_df.x, SF_Crime_df.y), crs='EPSG:4269')
SF_data_points

SF_shape = "cb_2018_06_tract_500k.shp"  # load California state code "06" for LA shape file 

SF_map = gpd.read_file(SF_shape)

SF_points_in_map = gpd.sjoin(SF_data_points.to_crs('EPSG:4269'), 
                         SF_map, 
                         how='inner', 
                         op='within')  ## doing inner because the census data is for whole of California and only LA crime data census tracts
print("I am done")
```
